```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig ,ax = plt.subplots(nrows=1, ncols=3, figsize=(16,6), sharex=True, sharey=True)
#Membuat figure
ax[0].plot(x, tan, color='blue', label='tanx')
ax[1].plot(x, sin, color='red', label ='sinx')
ax[2].plot(x, cos, color='green', label='cosx')
fig.legend(loc='upper center')


#Memberi Title
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(16,10))
ax.plot(x,y)
ax.set_title('Grafik cosX')

#Membuat yang kotak kecil
aq1 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
aq1.set_title('Grafik CosX^5')
aq1.plot(x, y, color="b")
aq1.plot(x, y5, color="y")

aq2 = ax.inset_axes([-0.1, -0.2, 0.3, 0.3])
aq2.set_title('Grafik CosX^2')
aq2.plot(x, y, color="b")
aq2.plot(x, y2, color="r")

aq3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
aq3.set_title('Grafik CosX^4')
aq3.plot(x, y, color="b")
aq3.plot(x, y4, color="k")

aq4 = ax.inset_axes([0.9, -0.2, 0.3, 0.3])
aq4.set_title('Grafik CosX^3')
aq4.plot(x, y, color="b")
aq4.plot(x, y3, color="g")
```
