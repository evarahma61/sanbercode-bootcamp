note : warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2

fig = plt.figure(figsize=(10,8))
ax = fig.subplots()
ax.plot(x, cos, color='red', label='Cosx')
ax.plot(x, cos3, linestyle='--', color='#228922', label='3 Cosx')
ax.plot(x, sin, linestyle='dashdot', color='#0000FF', label='Sin(x-90)')
ax.plot(x, sin2,'x', color='#FFD700', label='2 Sinx')
ax.legend(loc='center')
plt.show()

ax.set_xlabel('Z-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
```
